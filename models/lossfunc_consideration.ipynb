```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
margin = 2
x = torch.linspace(0,10,1000)
normal = F.relu(margin-x)
relu_log = F.relu(-torch.log(x))
relu_log_app = torch.log(1+margin/x)

plt.plot(x,normal.detach().numpy(),label="normal")
plt.plot(x,relu_log.detach().numpy(),label="normal")
plt.plot(x,relu_log_app.detach().numpy(),label="normal")
plt.legend()
plt.show()
```

```python
# normal
x = torch.linspace(0,10,1000)
plus = F.relu(x)
minus = F.relu(margin-x) #equal to relu_log
plt.figure()
plt.plot(x,plus,label="log_plus")
plt.plot(x,minus,label="log_minus")
plt.legend()
plt.show()
```

```python
# relu_log
x = torch.linspace(0,10,1000)
plus = F.relu(torch.log(x/margin))
minus = F.relu(-torch.log(x/margin))
plt.figure()
plt.plot(x,plus,label="log_plus")
plt.plot(x,minus,label="log_minus")
plt.legend()
plt.show()
```

```python
x = torch.linspace(0,10,1000)
plus = F.relu(torch.log(x/margin))
minus = F.relu(-torch.log(x/margin)) #equal to relu_log
x = torch.log(x/margin)
plt.figure()
plt.plot(x,plus,label="log_plus")
plt.plot(x,minus,label="log_minus")
plt.legend()
plt.show()
```

```python
x = torch.linspace(0,10,1000)
plus = torch.log(1+x/margin)
minus = torch.log(1+margin/x) #equal to relu_log
plt.figure()
plt.plot(x,plus,label="log_plus")
plt.plot(x,minus,label="log_minus")
plt.legend()
plt.show()
```
